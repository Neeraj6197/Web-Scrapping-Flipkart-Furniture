## Showcase:

```python
showcase_df=pd.DataFrame()
for j in range (0,33):
    web_page = requests.get("https://www.flipkart.com/furniture/showcases/pr?sid=wwe%2C3sa&otracker=categorytree&page={}".format(j)).text
    bs = BeautifulSoup(web_page,"html.parser")
    furnitures = bs.find_all('div',class_='_4ddWXP')

    
    name=[]
    discounted_price=[]
    original_price=[]

    for i in furnitures:
        name.append(i.find('a',class_='s1Q9rs').text)
        discounted_price.append(i.find(class_='_30jeq3').text)
        try:
            original_price.append(i.find(class_='_3I9_wc').text)
        except:
            original_price.append(i.find(class_='_30jeq3').text)
    #fetching product details of each furniture:
    anchors = bs.find_all('a',class_='_2rpwqI')
    all_links = list()

    for link in anchors:
        if(link.get('href') != '#'):
            linktext = "https://www.flipkart.com" + link.get('href')
            all_links.append(link)
        
    product_details = []
    for i in all_links:
        linktext1 = "https://www.flipkart.com" + i.get('href')
        linkpage = requests.get(linktext1).text
        lb = BeautifulSoup(linkpage,'html.parser')
        details = lb.find_all('li',class_='_21Ahn-')
        product_details.append(details)
        
    d = {'furniture_type':'Showcase','name':name,'discounted_price':discounted_price,'original_price':original_price,'product_details':product_details} 

    df = pd.DataFrame(d)
    showcase_df = showcase_df.append(df, ignore_index=True)
```
